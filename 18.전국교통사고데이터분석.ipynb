```python
df=pd.read_csv("C://Users//whs38//OneDrive//바탕 화면//사이버연수원//전국교통사고.csv",encoding='euc-kr',index_col=0)
df
```

```python
#1.서울에서 일어난 요일별 사망자수, 사장자수를 출력하시오.
df[df["발생지시도"]=="서울"].groupby("요일")[["사망자수","사상자수"]].sum()
```

```python
#2.서울 강북구 지역의 교통사고 위치를 지도로 표시하고 각마커에는 법규위반내용을 표시하시오.
kangbuk=df[(df["발생지시도"]=="서울") & (df["발생지시군구"]=="강북구")][["법규위반","위도","경도"]]

import folium
map=folium.Map(location=(kangbuk["위도"].mean(),kangbuk["경도"].mean()))
for i,s in kangbuk.iterrows():
    marker=folium.Marker(location=(s["위도"],s["경도"]),popup=s["법규위반"])
    marker.add_to(map)
map
map.save("kangbuk.html")
```

```python
#3.부산에서 발생한 총 사망자수를 구하시오.
df[df["발생지시도"]=="부산"][["사망자수"]].sum()
```

```python
#4.서울지역 월요일에 발생한 사상자 수, 사고유형,# 법규위반을 구하시오.
df[(df["발생지시도"]=="서울") & (df["요일"]=="월")][["사상자수","사고유형","법규위반"]]
```
