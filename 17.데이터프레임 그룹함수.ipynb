```python
df=pd.read_excel("C://Users//whs38//OneDrive//바탕 화면//사이버연수원//hotel.xlsx",index_col=0)
df
```

```python
#등급별 가격의 합
df.groupby("grade")[["price"]].sum()
```

```python
#등급별 가격의 평균
df.groupby("grade")[["price"]].mean()
```

```python
df1=df.groupby("grade")[["price"]].agg(['sum','mean'])
df1.columns=["가격의 합","가격의 평균"]
df1
```

```python
def fn(v):
    print("v=",v)
    print("===============")
    return v.sum()
```

```python
df.groupby("grade")["price"].agg(fn)
```

```python
def fn1(v):
    if v.mean()>8000:
        return "높다"
    else:
        return "낮다"
df.groupby("grade")["price"].agg(fn1)
```

```python
df.groupby("grade")["price"].agg(lambda v: v.mean())
```
